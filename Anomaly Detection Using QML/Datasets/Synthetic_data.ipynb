```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Normal data (e.g., normal traffic durations)
normal_data = np.random.normal(loc=2, scale=0.5, size=100)  # mean=2s, std=0.5

# Anomalies (e.g., unusual long or short durations)
anomalies = np.random.uniform(low=5, high=7, size=10)  # far from normal mean

# Combine
data = np.concatenate([normal_data, anomalies])
labels = np.concatenate([np.zeros(len(normal_data)), np.ones(len(anomalies))])  # 0 = normal, 1 = anomaly

# Shuffle
indices = np.arange(len(data))
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

# Create DataFrame
df = pd.DataFrame({
    'duration': data,
    'label': labels
})

# Plot
plt.scatter(np.arange(len(df)), df['duration'], c=df['label'], cmap='coolwarm')
plt.title("Synthetic Univariate Dataset (Duration vs Index)")
plt.xlabel("Index")
plt.ylabel("Duration")
plt.show()

# Save to CSV
df.to_csv("synthetic_cyber_dataset.csv", index=False) 
```
